```python
#Atlet doğum yılı yerine direkt yaşlarına bakmak daha kolay olur diye yaş sütunu ekledim
df2['Athlete Age'] = 2020 - df2['Athlete year of birth']
df2
```

```python
df2 = df2.drop(columns=['Athlete year of birth', 'Athlete club', 'Athlete age category'])
```

```python
df2 = df2.dropna(subset=["Athlete Age"])
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.replace(' h', '')
df2.head()
```

```python
df2['Athlete Age'] = df2['Athlete Age'].astype(int)
```

```python
df2 = df2.reset_index(drop=True)
df2.head()
```

```python
gender_counts = df2['Athlete gender'].value_counts()

plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Athlete Gender')
for i in range(len(gender_counts)):
    plt.text(i, gender_counts.values[i], str(gender_counts.values[i]), ha='center', va='bottom')

plt.show()
```

```python
yearly_participants = df['Year of event'].value_counts().sort_index()
plt.plot(yearly_participants.index, yearly_participants.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.title('Number of Participants by Year')
plt.show()
```

```python
start_year = 2017
filtered_df = df[df['Year of event'] >= start_year]
yearly_participants = filtered_df['Year of event'].value_counts().sort_index()
plt.plot(yearly_participants.index, yearly_participants.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.title('Number of Participants by Year')
for i, txt in enumerate(yearly_participants.values):
    plt.text(yearly_participants.index[i], txt, str(txt), ha='right', va='bottom')
plt.show()
```

```python
country_counts = df2['Athlete country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='viridis')
plt.xlabel('Country')
plt.ylabel('Number of Participants')
plt.title('Top 10 Countries by Number of Participants')
plt.xticks(rotation=45)
plt.show()
```

```python
# 'Athlete performance' sütununu formatı uymadığı için dönüştürmem gerekti
df2['Athlete performance'] = pd.to_timedelta(df2['Athlete performance']).dt.total_seconds() / 3600

# Yaşa göre gruplama ve ortalama zaman hesaplama
age_avg_time = df2.groupby('Athlete Age')['Athlete performance'].mean()

plt.figure(figsize=(10, 6))
plt.plot(age_avg_time.index, age_avg_time.values, marker='o', linestyle='-')
plt.xlabel('Athlete Age')
plt.ylabel('Average Performance (hours)')
plt.title('Average Performance by Athlete Age')
plt.grid(True)
plt.show()
```
